```python
# Featured Selection Part
```

```python
movies.columns
```

```python
# id, title,genre, overview
movie = movies[["id","title","overview","genre"]]
movie
```

```python
movie['tags'] = movie['overview']+movie['genre']
```

```python
movie
```

```python
new_data = movie.drop(columns = ['overview','genre'])
```

```python
new_data
```

```python
# tag column convert into vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"].index[0]
```

```python
new_data.info()
```

```python
list(enumerate(similarity[2]))
```

```python
distance = sorted(list(enumerate(similarity[2])) ,reverse=True,key= lambda vector: vector[1] )
```

```python
distance
```

```python
for i in distance[0:5]:
    print(i)
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]])
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recomand(movie):
    index = new_data[new_data['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])) ,reverse=True,key= lambda vector: vector[1] )
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recomand("Iron Man")
```
